# **1. What is the average selling price of cars for each dealer, and how does it compare across different dealers?**

```python
average_selling_price = df.groupby('Dealer_Name')['Annual Income'].mean()
print(average_selling_price)

# Visualization for comparison
plt.figure(figsize=(24, 6), facecolor = 'y')
sns.barplot(x=average_selling_price.index, y=average_selling_price.values)
plt.xlabel('Dealer_Name', fontsize = 45)
plt.ylabel('Average Selling Price', fontsize = 30)
plt.title('Average Selling Price by Dealer')
plt.xticks(rotation=80, fontsize = 14)
plt.yticks(fontsize = 14)
plt.show()
```

# **2. Which car brand (Company) has the highest variation in prices, and what does this tell us about the pricing trends?**

```python
price_variation = df.groupby('Company')['Price ($)'].std()
highest_variation_brand = price_variation.idxmax()
print(f"The car brand with the highest price variation is: {highest_variation_brand}")

#Further analysis on pricing trends
print(f"Price variation for each company:\n{price_variation}")

#Visualization for better understanding
plt.figure(figsize=(12, 6))
sns.barplot(x=price_variation.index, y=price_variation.values)
plt.xlabel("Car Company")
plt.ylabel("Price Standard Deviation")
plt.title("Price Variation Across Car Brands")
plt.xticks(rotation=45)
plt.show()
```

# **3. What is the distribution of car prices for each transmission type, and how do the interquartile ranges compare?**

```python
# Calculate the distribution of car prices for each transmission type
transmission_price_distribution = df.groupby('Transmission')['Price ($)'].describe()
print(transmission_price_distribution)

# Calculate and compare interquartile ranges (IQR)
iqr_by_transmission = df.groupby('Transmission')['Price ($)'].apply(lambda x: x.quantile(0.75) - x.quantile(0.25))
print("\nInterquartile Ranges (IQR) by Transmission Type:\n", iqr_by_transmission)


# Visualization for better understanding
plt.figure(figsize=(10, 6))
sns.boxplot(x='Transmission', y='Price ($)', data=df)
plt.title('Distribution of Car Prices by Transmission')
plt.xlabel('Transmission')
plt.ylabel('Price ($)')
plt.show()
```

# **11. How does car price vary with engine size across different car colors, and which colors have the highest price variation?**

```python
# Group data by color and engine size, then calculate the average price
color_engine_price = df.groupby(['Color', 'Engine'])['Price ($)'].mean().unstack()

# Calculate the price variation (standard deviation) for each color
price_variation_by_color = color_engine_price.std(axis=1)

# Find the colors with the highest price variation
colors_highest_variation = price_variation_by_color.nlargest(5)  # Get top 5

print("Price variation by color:\n", price_variation_by_color)
print("\nColors with the highest price variation:\n", colors_highest_variation)

# Visualization (optional):
plt.figure(figsize=(12, 6))
sns.heatmap(color_engine_price, annot=True, fmt=".0f", cmap="viridis")
plt.title('Average Car Price by Color and Engine Size')
plt.xlabel('Engine Size')
plt.ylabel('Color')
plt.show()

plt.figure(figsize=(8,6))
sns.barplot(x=price_variation_by_color.index, y=price_variation_by_color.values)
plt.title("Price Variation Across Car Colors")
plt.xlabel("Car Color", fontsize = 25)
plt.ylabel("Price Standard Deviation", fontsize = 25)
plt.xticks(rotation=45)
plt.show()
```

# **12. Is there any seasonal trend in car sales based on the date of sale?**

```python
# Convert 'Date' to datetime objects
df['Date'] = pd.to_datetime(df['Date'])

# Extract month and year
df['Month'] = df['Date'].dt.month
df['Year'] = df['Date'].dt.year

# Group by month and year, count sales
monthly_sales = df.groupby(['Year', 'Month'])['Price ($)'].count().reset_index()

# Create a time series
monthly_sales['Date'] = pd.to_datetime(monthly_sales['Year'].astype(str) + '-' + monthly_sales['Month'].astype(str), format='%Y-%m')
monthly_sales = monthly_sales.set_index('Date')

# Plot the time series
plt.figure(figsize=(12, 6))
plt.plot(monthly_sales['Price ($)'])
plt.title('Monthly Car Sales Trend')
plt.xlabel('Date', fontsize = 25)
plt.ylabel('Number of Cars Sold', fontsize = 25)
plt.xticks(rotation=45)
plt.grid(True)
plt.show()


#Further analysis (optional):  decompose the time series to identify trends, seasonality, and residuals.

from statsmodels.tsa.seasonal import seasonal_decompose

decomposition = seasonal_decompose(monthly_sales['Price ($)'], model='additive') # or 'multiplicative'

plt.figure(figsize=(12, 8))
decomposition.plot()
plt.show()
```
